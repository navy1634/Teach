```python
import pandas as pd
import random

average_list = []

N = 1

for test in range(2,N):
  df = pd.DataFrame(columns= answer.keys())

  # DataFrameの作成

  for i in range(1, test):
    name = "student_" + str(i)
    ans_list = []
    for ans in answer.values():
      ans_list.append(random.choice(ans))
    df.loc[name] = ans_list

  test_N = 100
  question_N = 10

  count_test = 0
  for _ in range(test_N):
    df_copy = df.copy()
    question = {
    "grade": "あなたの学年は？　[1,2,3]",
    "class" : "あなたのクラスは？　[1,2]",
    # "number" : "あなたの出席番号は？ range(1,35)",
    "sex" : 'あなたの性別は？　["男", "女", "その他"]',
    "club" :' 所属する部活は？　["陸上","バスケ", "剣道","卓球", "ESS", "茶道","音楽","自然科学部","放送映像文化部","その他","未所属"]',
    "subject": '好きな教科は？　["国語", "数学", "理科","社会","英語"]',
    "door" :'あなたはどちら？　 ["アウトドア派", "インドア派"]',
    "bus" : 'バスの方向は？　["相生", "新宮", "上郡", "その他"]',
    "foods" :' どちらが好き？　["肉", "魚"]',
    "time" : 'どれが好き？　["朝", "昼", "夜"]',
    "kinotake" :'どちらが好き？　 ["きのこの山", "たけのこの里"]',
    "breakfast" :' どちらが好き？　["ごはん", "パン"]',
    "season" : 'どれが好き？　["春", "夏", "秋", "冬"]',
    "birth" : '誕生月は？　range(1,12)',
    "living" : '住んでいるところは？　["相生", "たつの", "赤穂", "姫路", "上郡", "宍粟", "太子","明石", "その他"]',
    "parents" : '親？　[0,1]',
    "concil" : '委員ですか？　[0,1]',
    "threes" : '三班ではどれ？　["A", "B", "C"]',
    "flower" : '花は好き？　[0,1]',
    "grassies" : '眼鏡してますか？　[0,1]',
    "science" : '理系ですか？　[0,1]',
    "apple" : '携帯はアップル？　[0,1]',
    "jogorstd" : 'どちら？　["運動", "勉強", "どっちも好き","どっちも嫌い"]',
    "asmr" : 'ASMRはすき？　[0,1]',
    "likedis" : '好き嫌いある？　[0,1]',
  }
    count = 0
    for _ in range(question_N):
      # 質問内容を選ぶ
      choice = random.choice(list(question.keys()))

      # 答える
      # ans = input(question[choice])

      ans = random.choice(answer[choice])

      # 数字だった時の例外処理
      # if ans.isdecimal():
      #   ans = int(ans)

      # print(ans)

      # 聞いた質問を省く
      del question[choice]

      df_copy = df_copy[df.loc[:,choice] == ans]
      count += 1

      # 生徒数が1以下なら終わり
      if len(df_copy) < 2:
        break

    count_test += count

  average = count_test / test_N
  average_list.append(average)
```
